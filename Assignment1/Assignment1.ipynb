```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.linspace(0.,1.,10)
X_valid = np.linspace(0.,1.,100)
np.random.seed(9248)
t_train = np.sin(4*np.pi*X_train) + 0.3 * np.random.randn(10)
t_valid = np.sin(4*np.pi*X_valid) + 0.3 * np.random.randn(100)
```

```python
plt.scatter(X_train[:,], t_train, color = "red")
plt.scatter(X_valid[:,], t_valid, color = "blue")
```

```python
N = len(X_train)
M = len(X_valid)

new_col = np.ones(N)

X = np.c_[np.ones(N), X_train, X_train**2,X_train**3,X_train**4,X_train**5,X_train**6,X_train**7,X_train**8,X_train**9]
```

```python
w = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,t_train))
print(w)
```

```python
y = np.dot(X,w)

diff_train = np.subtract(t_train,y)
err_train = np.dot(diff_train.T,diff_train)/N  #training error
print(err_train)
RMSE_train = np.sqrt(err_train)
print(RMSE_train)
```

```python
#lets do a test against the validation data
X_valid_for_test = np.c_[np.ones(M), X_valid, X_valid**2,X_valid**3,X_valid**4,X_valid**5,X_valid**6,X_valid**7,X_valid**8,X_valid**9]

y_valid = np.dot(X_valid_for_test,w)
```

```python
plt.scatter(X_train, t_train, color = 'magenta', label = 'training example')
plt.scatter(X_valid, t_valid, color = 'blue', label = 'test example')
plt.scatter(X_valid,y_valid , color = 'green', label = 'prediction')
plt.scatter(X_train,t_train , color = 'red', label = 'prediction')
plt.plot(X_valid,y_valid , color = 'green',linestyle='solid')
plt.legend()
plt.show()
```
